```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
data=pd.read_csv('E:/5th sem/python/Project/train.csv').as_matrix()
clf=DecisionTreeClassifier()
```

```python
x_train=data[0:21000,1:]
train_label=data[0:21000,0]
clf.fit(x_train,train_label)
```

```python
x_test=data[21000:,1:]
actual_label=data[21000:,0]
```

```python
d=x_test[8]
d.shape=(28,28)
plt.imshow(255-d,cmap='gray')
print(clf.predict( [x_test[8]] ))
plt.show()
```

```python
for j in range(0,10):
    d=x_test[j]
    d.shape=(28,28)
    plt.imshow(255-d,cmap='gray')
    print('predicted value:',clf.predict([x_test[j]]))
    plt.show()
```

```python
p=clf.predict(x_test)
```

```python
count=0
for i in range(0,21000):
    if p[i]==actual_label[i]:
        count+=1
    else:
        count=0
count=count/21000 * 100
print('Accuracy:',count)      #this accuracy cell should be deleted
```
